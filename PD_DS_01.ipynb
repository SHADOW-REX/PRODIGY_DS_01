```python
# Importing Libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data loading and preparation
data=pd.read_csv('Cost_of_Living_Index_by_Country_2024.csv')
data_sorted = data.sort_values(by='Cost of Living Index', ascending=False)
```

```python
top_10_expensive = data_sorted.head(10)
top_10_cheapest = data_sorted.tail(10)
print("Top 10 Most Expensive Countries:")
print(top_10_expensive[['Country', 'Cost of Living Index']])
print("\nTop 10 Cheapest Countries:")
print(top_10_cheapest[['Country', 'Cost of Living Index']])
```

```python
# Generate and reverse the color palettes
expensive_palette = sns.color_palette('flare', len(top_10_expensive))[::-1]
cheapest_palette = sns.color_palette('crest', len(top_10_cheapest))[::-1]
```

```python
# Visualization

plt.figure(figsize=(16, 8))

plt.suptitle('Comparison of Cost of Living Index: Most Expensive vs. Cheapest Countries', fontsize=16)

# Expensive Countries
# Expensive Countries
plt.subplot(1, 2, 1)
sns.barplot(x='Cost of Living Index', y='Country', hue='Country', data=top_10_expensive, palette=expensive_palette, legend=False)
plt.title('Top 10 Most Expensive Countries', fontsize=14)
plt.xlabel('Cost Of Living Index')
plt.ylabel('Countries')

# Cheapest Countries
plt.subplot(1, 2, 2)
sns.barplot(x='Cost of Living Index', y='Country', hue='Country', data=top_10_cheapest, palette=cheapest_palette, legend=False)
plt.title('Top 10 Cheapest Countries', fontsize=14)
plt.xlabel('Cost Of Living Index')
plt.ylabel('Countries')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Categorical data visualisation
categories = ['Groceries Index', 'Restaurant Price Index', 'Rent Index']
```

```python
# Bar plots for the top 10 most expensive countries
for category in categories:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=top_10_expensive, x=category, y='Country', hue='Country', palette=expensive_palette, legend=False)
    plt.title(f'{category} for the Top 10 Most Expensive Countries')
    plt.xlabel(category)
    plt.ylabel('Country')
    plt.show()
```

```python
# Bar plots for the top 10 most cheapest countries
for category in categories:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=top_10_cheapest, x=category, y='Country', hue='Country', palette=cheapest_palette, legend=False)
    plt.title(f'{category} for the Top 10 Most Cheapest Countries')
    plt.xlabel(category)
    plt.ylabel('Country')
    plt.show()
```
